```
# Part 1 Data Preprocessing

## Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

## Importing the dataset
df=pd.read_csv('/content/drive/My Drive/ANN (Krish Naik)/Churn_Modelling.csv')
df.head()
X=df.iloc[:,3:13]
y=df['Exited']

## Creating dummy variables
X=pd.get_dummies(X,columns=['Geography','Gender'],drop_first=True)

## Splitting the dataset into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)

## Feature Scaling
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)

# Part 2: Let's make ANN

## Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU,Flatten,BatchNormalization,Embedding,Activation
from keras.layers import Dropout

## Intialising ANN
classifier=Sequential()

## Adding the input layer and first hidden layer
classifier.add(Dense(units=6,kernel_initializer='he_normal',activation='relu',input_dim=11))

classifier.add(Dropout(0.3))

## Adding second hidden layer
classifier.add(Dense(units=20,kernel_initializer='he_normal',activation='relu'))

classifier.add(Dropout(0.4))

## Adding third hidden layer
classifier.add(Dense(units=15,kernel_initializer='he_normal',activation='relu'))

classifier.add(Dropout(0.2))

## Adding the output layer
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid')) ##output layer: sigmoid AF works better

## Compiling the ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

## Fitting the ANN to the training dataset

model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)

## Predicting the Test dataset results
y_pred=classifier.predict(X_test)
y_pred=(y_pred>0.5)

## Making the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)

## Calculate the accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print(score)

```
